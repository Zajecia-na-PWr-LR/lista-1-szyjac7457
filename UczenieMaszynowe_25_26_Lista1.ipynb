## Zadanie 3
utwórz ramkę danych pozwalającą odpowiedzieć na następujące pytania
- Jaka dziedzina naukowa (*Branch*) jest najbardziej stresująca?
- Czy mediana snu wśród studentów informatyki (*IT*) jest wyższa niż wśród studentów mechaniki (*Mechanical*)?
### UWAGA!
Uwzględnij wyłącznie informacje o pełnoletnich studentach.

```python
adults = data[data['Age'] >= 18]

result3 = adults.groupby('Branch').agg(
    Average_Stress_Level=('Stress_Level_1_to_10', 'mean'),
    Median_Sleep_Hours=('Sleep_Hours', 'median')
).reset_index()

result3 = result3.sort_values(by='Average_Stress_Level', ascending = False)
result3
```

## Zadanie 5:
Utwórz wykresy, które pozwalają przedstawić następujące informacje:
- Najpopularniejsze tematy filmów na YouTube Shorts w ostatnim kwartale 2024 roku
- Jaki hasztag najbardziej opłaca się umieszczać w  postach na poszczególnych platformach?

```python
social_data = pd.read_csv('/content/drive/MyDrive/social_media_viral_content_dataset.csv')

social_data['post_datetime'] = pd.to_datetime(social_data['post_datetime'])

q4_2024_shorts = social_data[
    (social_data['platform'] == 'YouTube Shorts') &
    (social_data['post_datetime'].dt.year == 2024) &
    (social_data['post_datetime'].dt.quarter == 4) &
    (social_data['content_type'] == 'video')
]

plt.figure(figsize=(10, 6))
sns.countplot(data=q4_2024_shorts, x = 'topic', order=q4_2024_shorts['topic'].value_counts().index)

social_data['single_hashtag'] = social_data['hashtags'].str.split()
exploded_data = social_data.explode('single_hashtag')

best_hashtags = exploded_data.groupby(['platform', 'single_hashtag'])['engagement_rate'].mean().reset_index()
best_hashtags = best_hashtags.loc[best_hashtags.groupby('platform')['engagement_rate'].idxmax()]

plt.figure(figsize=(12, 6))
sns.barplot(data=best_hashtags, x='engagement_rate', y='single_hashtag', hue='platform')
```
